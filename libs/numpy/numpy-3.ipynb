```python
np.eye(3, k = -1) # or just np.eye(3) ==> will give diagonal matrix.
eye_array = np.eye(3, k = 1)
eye_array[eye_array == 0] = 2
print(eye_array)
eye_array[:2] = 3 # will change first 2 rows
print(eye_array)
eye_array[:, 0] = 4 # will change just first column
print(eye_array, "\n")
```

```python
np.sort(eye_array)
np.sort(eye_array, axis=0) # sorting columns
np.sort(eye_array, axis=0, kind="mergesort") # can also give heapsort
```
