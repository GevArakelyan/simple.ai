# Global Interpreter Lock("GIL")
Simple implementation, Efficient for single-threaded code, No deadlocks(if you adhere to
locking discipline) Some parallelism by releasing GIL

## Parallelism in Python today
1. Parallelize functions internally (e.g. OpenMP)
    Operations need to be large

---
2. Release the GIL in C/C++
    Doesn't scale to many cores

---
3. Multiple processes (multiprocessing)
Challenges with communication between processes
Object need to be serialized
Requires extra copying
Incompatibility with C/C++ libraries
