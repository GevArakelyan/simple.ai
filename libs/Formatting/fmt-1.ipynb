```python
n = 10
print(f"The number is {n:e}.")
print(f"The number is {n:x}.")
print(f"The number is {n:o}.")
print(f"The number is {n:06}.")
print(f"The number is {n:6}.")
print(f"The number is {n:.2f}.")

n = 10000000
print(f"The number is {n:,.2f}.")
print(f"The number is {n:_.2f}.")

n = 0.345
print(f"The number is {n:%}")
print(f"The number is {n:.2%}")
print(f"The number is {n:.1%}")
```

```python
greet = "Hi"
print(f"{greet:>4}")
print(f"{greet:^4}")
print(f"{greet:<4}")

print(f"{greet:_>4}")
print(f"{greet:_^4}")
print(f"{greet:_<4}")
```

```python
import datetime

today = datetime.datetime.now()
print(f"now is: {today}")
print(f"now is: {today:%H:%M:%S.%f}")
print(f"now is: {today:%Y:%m:%d %H:%M:%S.%f}")
print(f"now is: {today:%y:%m:%d %H:%M:%S.%f}")
print(f"now is: {today:%D}")
print(f"now is: {today:%T}")
print(f"now is: {today:%W}")
print(f"now is: {today:%A}")
print(f"now is: {today:%A, %B %d, %Y}")
print(f"now is: {today:%X}") # locale adjusted time and date
print(f"now is: {today:%x}")
```

```python
from dataclasses import dataclass

@dataclass
class User1:
    first_name: str
    last_name: str

user = User1("James", "Bond")
print(f"hi is {user}")

# print calls __str__ method if it exist othervise __repr__
# dataclass will inject __repr__ into our class
# if we will add manually __str__ method then print will call it, but we can call __repr__ version if do below

@dataclass
class User2:
    first_name: str
    last_name: str

    def __str__(self) -> str:
        return f"{self.first_name} {self.last_name}"

user2 = User2("James", "Bond")
print(f"he is {user2}")
print(f"he is {user2!r}")


```
