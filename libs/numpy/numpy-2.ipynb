```python
import numpy as np
a = np.array([1,2,3,4])

# grab first 3 elements
a[:3]
# grab last two elements
a[-2:]
# every other element
a[::2]

### For 2 dimentional array
firstRow = np.array([0,1,2,3,4,5])
a = np.full((6,6),0)
a[0] = firstRow
a[1] = firstRow+10
a[2] = firstRow+20
a[3] = firstRow + 30
a[4] = firstRow + 40
a[5] = firstRow+50


a[0, 3:5]
a[4:, 4:]
a[:, 2]
a[2::2, ::2]
```
