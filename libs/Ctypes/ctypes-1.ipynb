```python
# Suppose we have libhelloWorld.so
import ctypes # allows to call functions for shared libraries

helloWorld = ctypes.cdll.LoadLibrary("libhelloWorld.so")
helloWorld.helloWorld()
```

```python
import ctypes
libc = ctypes.cdll.LoadLibrary("libc.so.6")
libc.time()

libc.printf(b"Hello World\n")
libc.printf(b"An int%d, a double %f", 1234, ctypes.c_double(3.14))

libc.printf.argtypes = [ctypes.c_char_p, ctypes.c_int, ctypes.c_double]
libc.printf(b"String '%s', Int %d, Double %f\n", b"Hello World", 2020, 1.4142)

libc.strchr(b"Hello World", ord("W"))
libc.strchr.restype = ctypes.c_char_p
libc.strchr(b"Hello World", ord("W"))

```
