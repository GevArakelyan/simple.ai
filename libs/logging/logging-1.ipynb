```python
import logging

def print_vs_logging():
    logging.debug("debug info")
    logging.info("just some info")
    logging.error("Error happended")


def main():
    level = logging.DEBUG
    fmt = '[%(levelname)s] %(asctime)s - %(message)s'
    logging.basicConfig(level=level, format=fmt)
    print_vs_logging()

main()

```
