```python
from dataclasses import dataclass
```

```python


@dataclass
class User:
    firts_name: str
    last_name: str

user = User("James", "Bond")
print(f"here is: {user}")
```

```python
@dataclass
class Vector:
    x: float
    y: float

    def __truediv__(self, other: float):
        return Vector(self.x/other, self.y/other)

point = Vector(1,2)
print(point/2)
```
